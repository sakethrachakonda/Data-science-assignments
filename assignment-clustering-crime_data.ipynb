```python
#removing the outliers in Rape
import numpy as np
for i in df['Rape']:
    q1 = np.quantile(df.Rape,0.25)
    q3 = np.quantile(df.Rape,0.75)
    med = np.median(df.Rape)
    iqr = q3 - q1
    upper_bound = q3+(1.5*iqr)
    lower_bound = q1-(1.5*iqr)
    if i > upper_bound or i < lower_bound:
        df['Rape'] = df['Rape'].replace(i, np.median(df['Rape']))
sns.boxplot(df['Rape'])
plt.title('Rape after median imputation')
plt.show()
```

```python
X=df.iloc[:,0:]
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
pd.DataFrame(SS_X)
```

```python
#forming a group using clusters
#dbscan
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(SS_X)
dbscan.labels_
df['clusters']=dbscan.labels_
df
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['UrbanPop'], c=dbscan.labels_) 
```

```python
#agglomarative 
from sklearn.cluster import AgglomerativeClustering
agglomeartive=AgglomerativeClustering(n_clusters=4)
labels_agglo=agglomeartive.fit_predict(SS_X)
labels_agglo

df['clusters']=labels_agglo
df
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['UrbanPop'], c=labels_agglo) 
```

```python
#k-means
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(SS_X)
labels_kmeans=kmeans.labels_

df['clusters']=kmeans.labels_
df
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['UrbanPop'], c=kmeans.labels_) 
```
