```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
def plot_cloud(wordcloud):
    plt.figure(figsize=(15,30))
    plt.imshow(wordcloud)
    plt.axis("off");
```

```python
stopwords=STOPWORDS
```

```python
stopwords.add('will')
stopwords.add('amazon')
stopwords.add('\n')
stopwords.add('\r')
stopwords.add('online')
stopwords.add('shoe')
stopwords.add('if')
stopwords.add('nike')
stopwords.add('color')
stopwords.add('used')
stopwords.add('shoe')
```

```python
wordcloud=WordCloud(width=2000,height=1000,background_color='black',max_words=100,
                    colormap='Set2',stopwords=stopwords).generate(text)
plot_cloud(wordcloud)
```

```python
import string 
no_punc_text=text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(no_punc_text)
len(text_tokens)
print(text_tokens[0:60])
```

```python
my_stop_words=stopwords.words('english')
my_stop_words.append('the')
my_stop_words.append('as')
my_stop_words.append('in')
```

```python
len(my_stop_words)
```

```python
no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
no_stop_tokens
len(no_stop_tokens)
print(no_stop_tokens[0:50])
```

```python
#normalize the data
lower_words=[x.lower() for x in no_stop_tokens]
print(lower_words[0:50])
```

```python
doc = ' '.join(lower_words)
print(doc)
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
Lemmatizer=WordNetLemmatizer()
```

```python
lemmas=[]
for token in doc.split():
    lemmas.append(Lemmatizer.lemmatize(token))
    
print(lemmas)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(lemmas)
X
pd.DataFrame.from_records([vectorizer.vocabulary_]).T.sort_values(0,ascending=True).head(30)

```

```python
#bigrams and trigams
vectorizer_ngram_range=CountVectorizer(analyzer='word',ngram_range=(1,1),max_features=100)
bow_matrix_ngram=vectorizer_ngram_range.fit_transform(book)
bow_matrix_ngram
feature_names = vectorizer.get_feature_names_out()

print(vectorizer_ngram_range.get_feature_names_out())
print(bow_matrix_ngram.toarray())
```
