```python
X=df.iloc[:,0:]
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
pd.DataFrame(SS_X)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(SS_X,'complete'))
```

```python
#forming a group using clusters
#dbscan
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(SS_X)
dbscan.labels_
df['clusters']=dbscan.labels_
df
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['Balance'], c=dbscan.labels_)
```

```python
#agglomarative 
from sklearn.cluster import AgglomerativeClustering
agglomeartive=AgglomerativeClustering(n_clusters=4)
labels_agglo=agglomeartive.fit_predict(SS_X)
labels_agglo

df['clusters']=labels_agglo
df
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['Balance'], c=labels_agglo)
```

```python
#Elbow plot for means 
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
data=df.iloc[:,0:14]
pca=PCA(n_components=3)
pca_result=pca.fit_transform(data)
df_pca=pd.DataFrame(data=pca_result,columns=['PC1','PC2','PC3'])
inertia=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(pca_result)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(8,5))
plt.plot(range(1,11),inertia,marker='o')
plt.title('Elbow Plot for KMeans')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia(Within-cluster sum of squares)')
plt.show()
```
